```python
class present_step:
    def __init__(self,present_step,umbrella,required_times,time):
        self.present_step = present_step
        self.umbrella = umbrella
        self.required_times = required_times
        self.time = time 
    
    def goalTest(self):
        goal = ['R','R','R','R']
        return self.present_step == goal

    def moveGen(self):
        steps = []
        for i in range(len(self.present_step)):
            for j in range(len(self.present_step)):
                # umbrella's position is updated Rith this
                neR_position = 'R' if self.umbrella == 'L' else 'L'
                # neR_position = 1 if self.umbrella == 0 else 0

                if i == j and self.umbrella == self.present_step[i]:
                    neR_step = self.present_step.copy()
                    neR_step[i] = neR_position
                    steps.append(present_step(neR_step, neR_position, self.required_times, self.time +self.required_times[i]))
                elif self.umbrella == self.present_step[i] == self.present_step[j]:
                    neR_step = self.present_step.copy()
                    neR_step[i] = neR_step[j] = neR_position
                    steps.append(present_step(neR_step, neR_position, self.required_times, self.time +max(self.required_times[i],self.required_times[j])))
                
        return steps
    
    def __eq__(self,other):
        return isinstance(other, present_step) and self.present_step == other.present_step and self.umbrella == other.umbrella
    
    def __str__(self):
        return f' [ {self.present_step[0]}, {self.present_step[1]}, {self.present_step[2]}, {self.present_step[3]} ] , position of umbrella : {self.umbrella} , time taken to pass bridge : {self.time }'

    def __hash__(self):
        return hash(tuple(self.present_step))
```

```python
def filtersteps(children,open,closed):
    open_states = [o for o,parent in open]
    closed_states = [c for c,parent in closed]
    new_states = [state for state in children if state not in open_states and state not in closed_states and state.time<=60]
    return new_states

def constructStep(pair,closed):
    path = []
    parent_map = {}
    
    for node,parent in closed:
        parent_map[node] = parent
    
    node,parent = pair
    path.append(node)
    while parent:
        path.append(parent)
        parent = parent_map[parent]
    
    path.reverse()
    return path

def printStep(path):
    for i in range(len(path)):
        print(path[i])

def BFS(start):
    open = [(start,None)]
    closed = []
    while open:
        pair = open.pop(0)
        n,parent = pair
        if n.goalTest():
            path = constructStep(pair,closed)
            printStep(path)
            return
        closed.append(pair)
        children = n.moveGen()
        updated_step = filtersteps(children,open,closed)
        new_states = [(state,n) for state in updated_step]
        open = open + new_states
    return []

def DFS(start):
    open = [(start,None)]
    closed = []
    while open:
        pair = open.pop(0)
        n,parent = pair
        if n.goalTest():
            path = constructStep(pair,closed)
            printStep(path)
            return
        closed.append(pair)
        children = n.moveGen()
        updated_step = filtersteps(children,open,closed)
        new_states = [(state,n) for state in updated_step]
        open = new_states + open
    return []

present_step1 = present_step(['L','L','L','L'],'L',[5,10,20,25],0)
print("Solution using BFS Algorithm :")
BFS(present_step1)

print()

print("Solution using DFS Algorithm :")
DFS(present_step1)
```
