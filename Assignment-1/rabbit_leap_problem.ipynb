```python
class present_step:
  def __init__(self,present_step):
    self.present_step=present_step

  def goalTest(self):
    # present_step = ['w','w','w','-','e','e','e']
    # present_step2 = ['w','w','w','-','e','e','e']
    # print(present_step==present_step2) True
    # goal = [2,2,2,0,1,1,1]
    goal = ['W','W','W','_','E','E','E']
    return self.present_step == goal

  def moveGen(self):
    steps = []
    for i in range(len(self.present_step)):
      if self.present_step[i]=='E':
        if i+1<len(self.present_step) and self.present_step[i+1]=='_':
          new_step = self.present_step.copy()
          new_step[i],new_step[i+1] = '_','E'
          steps.append(present_step(new_step))
        elif i+2<len(self.present_step) and self.present_step[i+2]=='_':
          new_step=self.present_step.copy()
          new_step[i],new_step[i+2] = '_','E'
          steps.append(present_step(new_step))
      elif self.present_step[i]=='W':
        if i-1>=0 and self.present_step[i-1]=='_':
          new_step = self.present_step.copy()
          new_step[i],new_step[i-1] = '_','W'
          steps.append(present_step(new_step))
        elif i-2>=0 and self.present_step[i-2]=='_':
          new_step = self.present_step.copy()
          new_step[i],new_step[i-2] = '_','W'
          steps.append(present_step(new_step))
    return steps

  def __eq__(self,other):
    return self.present_step==other.present_step

  def __str__(self):
    return f'[ {self.present_step[0]}, {self.present_step[1]}, {self.present_step[2]}, {self.present_step[3]}, {self.present_step[4]}, {self.present_step[5]}, {self.present_step[6]} ]'

  def __hash__(self):
    return hash(tuple(self.present_step))

present_step1 = present_step([1,1,1,0,2,2,2])
# print(present_step1)
# print(present_step2)

steps = present_step1.moveGen()
for i in steps:
  print(i)
```

```python
def removeSeen(steps,open,closed):
  open_states = [o for o,parent in open]
  closed_states = [c for c,parent in closed]
  new_states = [new_state for new_state in steps if new_state not in open_states and new_state not in closed_states]
  return new_states

def constructStep(pair,closed):
  path = []
  parent_map = {}

  for node,parent in closed:
    parent_map[node] = parent

  node,parent = pair
  path.append(node)
  while parent:
    path.append(parent)
    parent=parent_map[parent]
  path.reverse()
  return path

def printSteps(path):
  for i in range(len(path)):
    print(path[i])

def bfs(start):
  open = [(start,None)]
  closed = []
  while open:
    pair = open.pop(0)
    n,parent = pair

    # print(n)
    if n.goalTest():
      path = constructStep(pair,closed)
      printSteps(path)
      return
    else:
      closed.append(pair)
      children = n.moveGen()
      updated_children = removeSeen(children,open,closed)
      new_children = [(state,n) for state in updated_children]
      open = open + new_children
  return []

def dfs(start):
  open = [(start,None)]
  closed = []
  while open:
    pair = open.pop(0)
    n,parent = pair
    if n.goalTest():
      path = constructStep(pair,closed)
      printSteps(path)
      return
    closed.append(pair)
    children = n.moveGen()
    updated_children = removeSeen(children,open,closed)
    new_children = [(state,n) for state in updated_children]
    open = new_children + open
  return []


present_step1 = present_step(['E','E','E','_','W','W','W'])
print("Solution using BFS Algorithm: ")
bfs(present_step1)

print()

print("Solution using DFS Algorithm: ")
dfs(present_step1)
```
